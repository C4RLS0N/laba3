```python
from deeppavlov import configs, build_model 

from razdel import sentenize
model = build_model(configs.ner.ner_ontonotes_bert_mult, download=True)
```

```python
inpt1 = int(input("Начало: "))
inpt2 = int(input("Конец: "))
inpt1=inpt1-1;


 

 

text = ''' Встретим пришельца лампадкой, Тихим и верным огнем. Только ни вздоха украдкой, Ни вздоха о нем! Яркого света не надо, Лампу совсем привернем. Только о лучшем ни взгляда, Ни взгляда о нем! Пусть в треволненье беспечном Год нам покажется днем! Только ни мысли о вечном, Ни мысли о нем! Станем «сестричками» снова, Крепче друг к другу прильнем. Только о прошлом ни слова, Ни слова о нем!  '''
 

x = list(sentenize(text))
 

print('\n Текст без обработки: \n' ,text, '\n' , '\n Первая нейросеть (natasha) (Разделение на предложения): \n' , x ,'\n')
print('Вторая нейросеть (deeppavlov) (Вторая нейросеть обработала результат первой (именованные сущности в тексте и классификации их по типам)): \n')
model([_.text for _ in x[inpt1:inpt2:1]])
```
